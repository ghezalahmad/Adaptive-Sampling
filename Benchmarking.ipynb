## Random Sampling

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random

# Load and preprocess the data
data = pd.read_excel("open_crosion.xlsx")
experiment_17_data = data[data['Experiment'] == 18]

# Adjust the coordinates to fit the grid
experiment_17_data['X-Coordinate_Adjusted'] = (experiment_17_data['X-Coordinate'] / 10).round().astype(int)
experiment_17_data['Y-Coordinate_Adjusted'] = (experiment_17_data['Y-Coordinate'] / 10).round().astype(int)

# Function to perform random sampling with visualization
def random_sampling_with_visualization(data, max_iterations=1000):
    sampled_points = []  # List to store sampled points
    corrosion_found = False
    
    for iteration in range(max_iterations):
        # Randomly select a point
        random_index = random.choice(data.index)
        random_point = data.loc[random_index]
        sampled_points.append((random_point['X-Coordinate_Adjusted'], random_point['Y-Coordinate_Adjusted']))
        
        # Check if the selected point is a corrosion point
        if random_point['Label'] == 1:
            print(f"Corrosion found at iteration {iteration + 1} at coordinates ({random_point['X-Coordinate_Adjusted'] * 10}cm, {random_point['Y-Coordinate_Adjusted'] * 10}cm)")
            corrosion_point = (random_point['X-Coordinate_Adjusted'], random_point['Y-Coordinate_Adjusted'])
            corrosion_found = True
            break
    
    if not corrosion_found:
        print(f"No corrosion found after {max_iterations} iterations.")
    
    # Visualization
    plt.figure(figsize=(10, 10))
    plt.grid(True)
    plt.title(f"Random Sampling in Experiment 17 - Iterations: {iteration + 1}")
    plt.xlabel("X-Coordinate (in adjusted units)")
    plt.ylabel("Y-Coordinate (in adjusted units)")

    # Plot all sampled points
    for point in sampled_points:
        plt.scatter(*point, color='blue', label='Sampled Point' if sampled_points.index(point) == 0 else "")

    # Highlight the corrosion point if found
    if corrosion_found:
        plt.scatter(*corrosion_point, color='red', label='Corrosion Point', s=100, edgecolors='black')

    plt.legend()
    plt.show()

    return iteration + 1  # Return the number of iterations required

# Perform random sampling with visualization on Experiment 17 data
iterations_required = random_sampling_with_visualization(experiment_17_data)
```

### Latine Hypercube Sampling

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import qmc

# Load and preprocess the data
data = pd.read_excel("open_crosion.xlsx")
experiment_17_data = data[data['Experiment'] == 17]

# Adjust the coordinates to fit the grid
experiment_17_data['X-Coordinate_Adjusted'] = (experiment_17_data['X-Coordinate'] / 10).round().astype(int)
experiment_17_data['Y-Coordinate_Adjusted'] = (experiment_17_data['Y-Coordinate'] / 10).round().astype(int)
n_samples_total = 100  # Total number of samples to generate

def iterative_lhs_corrosion_detection_and_plot(data, n_samples):
    """Iteratively perform Latin Hypercube Sampling to detect corrosion and plot the process."""
    sampler = qmc.LatinHypercube(d=2)  # Initialize the LHS sampler for 2 dimensions
    samples = sampler.random(n=n_samples)  # Generate LHS samples

    # Scale the samples to the grid boundaries
    samples[:, 0] = qmc.scale(samples[:, [0]], [0], [13]).flatten()  # Scale for X-coordinate
    samples[:, 1] = qmc.scale(samples[:, [1]], [0], [9]).flatten()  # Scale for Y-coordinate
    samples = np.round(samples)  # Round to nearest grid point

    corrosion_found = False
    iteration = 0

    plt.figure(figsize=(13, 9))
    plt.grid(True)
    plt.title("Latin Hypercube Sampling in Experiment 17")
    plt.xlabel("X-Coordinate (cm)")
    plt.ylabel("Y-Coordinate (cm)")

    # Plot all points in the grid as gray dots
    plt.scatter(data['X-Coordinate_Adjusted'] * 10, data['Y-Coordinate_Adjusted'] * 10, color='gray', label='Grid Points', alpha=0.5)

    for sample in samples:
        iteration += 1
        x, y = sample
        # Convert grid points to adjusted coordinates
        x_adj, y_adj = int(x), int(y)

        # Check if the sampled point is a corrosion point
        sample_data = data[(data['X-Coordinate_Adjusted'] == x_adj) & (data['Y-Coordinate_Adjusted'] == y_adj)]
        if not sample_data.empty and sample_data['Label'].iloc[0] == 1:
            corrosion_found = True
            plt.scatter(x_adj * 10, y_adj * 10, color='red', label='Corrosion Point' if iteration == 1 else "", s=100, edgecolors='black')
            print(f"Corrosion found at iteration {iteration} at grid coordinates ({x_adj}, {y_adj})")
            break
        else:
            plt.scatter(x_adj * 10, y_adj * 10, color='blue', label='Sampled Point' if iteration == 1 else "", alpha=0.7)

    plt.legend()
    plt.show()

    if not corrosion_found:
        print("No corrosion found within the sampled points.")

    return iteration, corrosion_found

# Perform iterative LHS and detect corrosion with plot
iterations_required, corrosion_found = iterative_lhs_corrosion_detection_and_plot(experiment_17_data, n_samples_total)

if corrosion_found:
    print(f"Corrosion was detected after {iterations_required} iterations.")
else:
    print("Corrosion was not detected within the specified number of iterations.")
```
